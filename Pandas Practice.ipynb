# Making changes to the data

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head()
```

```python
df = df.drop(columns=['Total'])
df.head()
```

```python
#Better adding of numbers
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df.head()
```

# Saving our Data (Exporting into Desired Format)

```python
df.to_csv('modified.csv', index=False)
```

# Filtering Data

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]

new_df.reset_index(drop=True, inplace=True) #No need to do new_df = new_df with inplace

new_df.head()
```

```python
df.loc[df['Name'].str.contains(' ')].head()
```

# Conditional Changes

```python
#For all Type 1 == Fire, Type 1 now = Flamer
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df.loc[df['Type 1'] == 'Flamer'].head()
```

```python
#Second parameter in loc can even be different than first
#This sets all dragon type pokemon as legendary
df.loc[(df['Type 1'] == 'Dragon') | (df['Type 2'] == 'Dragon'), 'Legendary'] = True
df.loc[(df['Type 1'] == 'Dragon') | (df['Type 2'] == 'Dragon')].head()
```

```python
df.loc[df['Total'] > 500, ['Generation','Legendary']] = 'TEST VALUE'
df.head()
```

```python
df.loc[df['Total'] > 500, ['Generation','Legendary']] = ['Test 1', 'Test 2']
df.head()
```

# Aggregate Statistics (Groupby)

```python
df = pd.read_csv('modified.csv')
df.head()
```

```python
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)
```

```python
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1']).count()
```

```python
df['count'] = 1
df.groupby(['Type 1', 'Type 2']).count()['count']
```
